# Symbolic calculations
This notebook contains all the symbolic calculations (using `SymPy`) that are needed to find the polynomial coefficients in both the ex-ante and the ex-post case in order to find candidate equilibria and candidate deviations.

We start by importing SymPy and defining a function that prints the polynomial coefficients as ready-to-use python assignments.

```python
from sympy import *

def print_coeffs(polynomial):
    coeffs = polynomial.all_coeffs()
    for i in range(0,len(coeffs)):
        print("coeffs[" + str(i) + "] = " + str(coeffs[i]).replace("atan(m)","np.arctan(m)"))
```

We define all the required SymPy symbols.

```python
a0,a,b,rho,lam,kap,thea,theb,the,bv,av,b0,m,bkap,akap,ai,bi,cbar,a_tilde,theta = symbols("a0 a b rho lam kap thea theb the bv av b0 m b^kappa a^kappa ai bi cbar a_tilde theta",positive=True)
```

## Ex-ante (nonpartisan) case
### Equilibrium candidates for group A

```python
f1 = (1+m**2) * a**2 + (1+m**2) * b0**2 + (2- 2 * m**2) *a *b0
f2 = a-a0
sub = 2 * (av-rho*bv) *m * kap * b0 * av /(the*atan(m))

polynomial = Poly(collect(expand(f1*f2-sub),a),a)
print_coeffs(polynomial)
```

### Deviation candidates for group A

```python
f1 = (1+m**2) * akap**2 + (1+m**2) * b0**2 + (2- 2 * m**2) *akap *b0
f2 = ai-a0
sub = 2 * (av-rho*bv) *m * kap * b0 * av /(the*atan(m))

polynomial = Poly(collect(expand(f1.subs(akap,kap*ai+(1-kap)*a)*f2-sub),ai),ai)
print_coeffs(polynomial)
```

### Equilibrium candidates for group B

```python
f1 = (1+m**2) * b**2 + (1+m**2) * a0**2 + (2- 2 * m**2) *a0 *b
f2 = b-b0
sub = 2 * (rho*bv-av) *m * kap * a0 * bv /(the*atan(m))

polynomial = Poly(collect(expand(f1*f2-sub),b),b)
print_coeffs(polynomial)
```

### Deviation candidates for group B

```python
f1 = (1+m**2) * bkap**2 + (1+m**2) * a0**2 + (2- 2 * m**2) *bkap *a0
f2 = bi-b0
sub = 2 * (rho*bv-av) *m * kap * a0 * bv /(the*atan(m))

polynomial = Poly(collect(expand(f1.subs(bkap,kap*bi+(1-kap)*b)*f2-sub),bi),bi)
print_coeffs(polynomial)
```

### Reverse search for group A equilibrium candidates
For the plot pointing out coordination problems for various $a_0$, we can also fix $a$ and solve the same equilibrium condition for $a_0$ to obtain equilibrium candidates.

```python
interaction = simplify((1+m**2) * a**2 + (1+m**2) * b0**2 + (2- 2 * m**2) * a * b0)
equation = Eq(interaction * (a-a0) - 2 * (av-rho*bv) * m * kap * b0 * av /(the*atan(m)),0)
closed_form_a0 = simplify(solve(equation,a0)[0])
print("Closed-form solution for a0:")
print(closed_form_a0)
```

## Ex-post (partisan) case
### Consistent strategy candidates for A

```python
f1 = (1+m**2) * b**2 + (1+m**2) * a**2 + (2- 2 * m**2) *b *a
f2 = a-a0
sub = 2*kap*m*b*av**2 / (atan(m) *thea)
polynomial = Poly(collect(expand(f1*f2-sub),a),a)
print("Polynomial in a:")
print_coeffs(polynomial)
```

### Reverse scan for A-consistent points
Fix $a$ and solve the same first-order condition for $b$.

```python
f1 = (1+m**2) * b**2 + (1+m**2) * a**2 + (2- 2 * m**2) * b * a
f2 = a-a0
sub = 2 * kap * m * b * av**2 / (atan(m) * thea)
polynomial = Poly(collect(expand(f1*f2-sub),b),b)
print("Polynomial in b (fixing a):")
print_coeffs(polynomial)
```

### Equation in $m$ for fixed $(a,b)$
This is the equation used in the reverse scan for the $m$-comparative-statics plot. It is not polynomial in $m$ because of the $m / \arctan(m)$ term, but the compact residual can still be derived symbolically.

```python
interaction = simplify((1+m**2) * (a**2 + b**2) + (2 - 2*m**2) * a * b)
compact_interaction = (a+b)**2 + m**2 * (a-b)**2
print("Difference from compact interaction form:")
print(simplify(interaction - compact_interaction))

residual_m = simplify(compact_interaction * (a-a0) - 2 * kap * m * b * av**2 / (atan(m) * thea))
print("Residual in m:")
print(residual_m)

print("Equivalent equation after multiplying by atan(m):")
print(expand(atan(m) * residual_m))
```

### Deviation candidates for A

```python
f1 = (1+m**2) * b**2 + (1+m**2) * akap**2 + (2- 2 * m**2) *b *akap
f2 = ai-a0
sub = 2 * kap * m * b * av**2 / (atan(m) *thea)
polynomial = Poly(collect(expand(f1.subs(akap,kap*ai+(1-kap)*a)*f2-sub),ai),ai)
print_coeffs(polynomial)
```

### Consistent strategy candidates for B

```python
f1 = (1+m**2) * b**2 + (1+m**2) * a**2 + (2- 2 * m**2) *b *a
f2 = b-b0
sub = 2 * kap * m * a * bv**2 * rho / (atan(m) *theb)
polynomial = Poly(collect(expand(f1*f2-sub),b),b)
print_coeffs(polynomial)
```

### Reverse scan for B-consistent points
Fix $b$ and solve the same first-order condition for $a$.

```python
f1 = (1+m**2) * b**2 + (1+m**2) * a**2 + (2- 2 * m**2) * b * a
f2 = b-b0
sub = 2 * kap * m * a * bv**2 * rho / (atan(m) * theb)
polynomial = Poly(collect(expand(f1*f2-sub),a),a)
print("Polynomial in a (fixing b):")
print_coeffs(polynomial)
```

### Deviation candidates for B

```python
f1 = (1+m**2) * a**2 + (1+m**2) * bkap**2 + (2- 2 * m**2) *a *bkap
f2 = bi-b0
sub = 2 * kap * m * a * bv**2 * rho / (atan(m) *theb)
polynomial = Poly(collect(expand(f1.subs(bkap,kap*bi+(1-kap)*b)*f2-sub),bi),bi)
print_coeffs(polynomial)
```

### Resultant for equilibrium candidates

In the ex-post case, the first-order condition for equilibrium candidates is a system of two polynomials in $a$ and $b$. In order to find solutions, we use the resultant method. It yields a degree $5$ polynomial in $b$ without a constant term (so one can obtain a degree $4$ polynomial by dividing by $b$). After determining the solutions to this polynomial, one can find the associated values for $a$.

```python
f1 = (1+m**2) * b**2 + (1+m**2) * a**2 + (2- 2 * m**2) *b *a
f2a = a-a0
f3a = 2 * kap * m * b * av**2 / (atan(m) * thea)

f2b = b-b0
f3b = 2 * kap * m * a * bv**2 * rho / (atan(m) *theb)

poly1 = Poly(expand(f1*f2a-f3a),(a,b))
poly2 = Poly(expand(f1*f2b-f3b),(a,b))

resu_poly = simplify(collect(resultant(poly1,poly2,a)/ b,b).subs(a0,0).subs(b0,0).subs(bv,1).subs(av,1))
print_coeffs(Poly(resu_poly,b))
sols = solve(resu_poly,b)
for sol in sols:
    print(sol)
```

## Ex-post (partisan) case, more general cost function
Here, we find expressions for the leading coefficients of the polynomials that describe deviation candidates in the setting similar to Herrera, Morelli, and Nunnari (2008).

### for A, candidates (general k)
For general k (cost function exponent), the polynomial is of degree k+2 in a. The expression uses $(a + a_0)^{k+1}$ to capture this. In the following code, `a` stands for $\tilde{a} = a - a_0$.

```python
# For general k, the structure is the same as k=1 - just with av^(k+1) in the constant term
# The key is: only coeffs[0], coeffs[1], coeffs[2], and coeffs[k+2] are non-zero
# See the k=15 demo cell below for illustration
expression = (1+m**2) * (a+a0)**2 + (1+ m**2) *b**2 + (2- 2*m**2)*(a+a0)*b
polynomial = Poly(collect(expand(expression),a),a)
print_coeffs(polynomial)
```

```python
# For general k, the polynomial degree is k+2
# The leading coefficient is always (m² + 1)
# The constant term involves av^(k+1)
# Pattern (only 4 non-zero coeffs):
# - coeffs[0] = m² + 1
# - coeffs[1] = 2*a0*m² + 2*a0 - 2*b*m² + 2*b
# - coeffs[2] = a0²*m² + a0² - 2*a0*b*m² + 2*a0*b + b²*m² + b²
# - coeffs[k+2] = -2*kap*m*b*av**(k+1)/(thea*atan(m))
# See k=15 demo below for illustration
```

**General pattern for arbitrary k:**
- Polynomial degree: k+2 (array length k+3)
- `coeffs[0] = m² + 1` (always)
- The coefficients for intermediate powers of a are 0
- Only `coeffs[1]`, `coeffs[2]`, and `coeffs[k+2]` are non-zero
- The constant term `coeffs[k+2] = -2κm·b·av^(k+1)/(θ·atan(m))`

```python
# Correct FOC for general k (derived from utility derivative)
# FOC: κ·m·2b·av^(k+1)/atan(m) = θ·a_tilde^k · ((1+m²)a² + (1+m²)b² + 2(1-m²)ab)
# where a = a0 + a_tilde

k_demo = 15
a = a0 + a_tilde  # Define a in terms of a0 and a_tilde

# FOC: RHS - LHS = 0, then divide by theta to match JavaScript pattern
LHS = (kap * m * 2 * b * av**(k_demo + 1)) / atan(m)
RHS = theta * a_tilde**k_demo * ((1+m**2)*a**2 + (1+m**2)*b**2 + 2*(1-m**2)*a*b)
foc = expand((RHS - LHS) / theta)

polynomial = Poly(foc, a_tilde)
coeffs = polynomial.all_coeffs()

print(f"For k={k_demo} (degree {polynomial.degree()}):")
print(f"Number of coefficients: {len(coeffs)}")
for i, c in enumerate(coeffs):
    power = polynomial.degree() - i
    print(f"  a_tilde^{power}: {c}")
```

### for A, in terms of $a^i$
The first coefficient corresponds to $(\tilde{a}^i)^{k+2}$. `a` stands for $a$, but `ai` stands for $\tilde{a}^i$.

```python
# Deviation candidates for A: FOC wrt ai (general k)
# Divide by theta to match JS pattern

k_demo = 15

akappa_expr = kap*(ai + a0) + (1-kap)*a
denom_akap = (1+m**2)*akappa_expr**2 + (1+m**2)*b**2 + 2*(1-m**2)*akappa_expr*b

# FOC: theta * ai^k * denom - kap * m * 2b * av^(k+1) / atan(m) = 0
# Divide by theta to match JS pattern
foc_dev = expand(ai**k_demo * denom_akap - kap * m * 2 * b * av**(k_demo + 1) / (theta * atan(m)))

polynomial_dev = Poly(foc_dev, ai)
coeffs_dev = polynomial_dev.all_coeffs()

print(f"Deviation candidates for k={k_demo} (degree {polynomial_dev.degree()}):")
print(f"Number of coefficients: {len(coeffs_dev)}")
for i, c in enumerate(coeffs_dev):
    power = polynomial_dev.degree() - i
    print(f"  ai^{power}: {simplify(c)}")
```

### Reverse scan for A (more general cost) - fix ã, solve for b
Fix ã = a - a0 and solve the same first-order condition for b.

```python
# Reverse scan for A: fix a (or equivalently a_tilde = a - a0), solve for b
# The equation is a quadratic in b

k_demo = 15
a = a0 + a_tilde

# Full polynomial without dividing by theta
A = (1+m**2) * theta * a_tilde**k_demo
B = 2*(1-m**2)*a*theta*a_tilde**k_demo - 2*kap*m*av**(k_demo+1)/atan(m)
C = (1+m**2)*theta*a_tilde**k_demo*a**2

# Build and simplify
foc_reverse = A * b**2 + B * b + C
foc_simplified = simplify(expand(foc_reverse))
print("Simplified polynomial:")
print(foc_simplified)

polynomial_rev = Poly(foc_simplified, b)
coeffs_rev = polynomial_rev.all_coeffs()

print(f"\nReverse scan (fix a, solve for b) for k={k_demo}:")
print(f"Degree in b: {polynomial_rev.degree()}")
print(f"Coefficients (highest to lowest power):")
for i, c in enumerate(coeffs_rev):
    power = polynomial_rev.degree() - i
    print(f"  b^{power}: {c}")
```
